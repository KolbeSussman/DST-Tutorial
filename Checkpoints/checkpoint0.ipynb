```python
# fizzbuzz
# you will need to use both iteration and control flow 
# go through all numbers from 1 to 30 in order
# if the number is a multiple of 3, print fizz
# if the number is a multiple of 5, print buzz
# if the number is a multiple of 3 and 5, print fizzbuzz and NOTHING ELSE
# if the number is neither a multiple of 3 nor a multiple of 5, print the number
def fizzbuzz(lst):
    for x in lst:
        if (x%3 == 0) & (x%5 == 0):
            print("fizzbuzz")
        elif x%3 == 0:
            print("fizz")
        elif x%5 == 0:
            print("buzz")
        else:
            print(x)
lst = []
for i in range(1, 31):
    lst.append(i)

fizzbuzz(lst)
```

```python
# load in the "starbucks.csv" dataset
# refer to how we read the cereal.csv dataset in the tutorial
import pandas as pd
df = pd.read_csv("../data/starbucks.csv")
df.head()
```

```python
df.reset_index(inplace = True)
```

```python
# output the calories, sugars, and protein columns only of every 40th row. 

def this_question(row):
    if (row['index']+1) % 40 == 0:
        output = {}
        output['calories'] = row['calories']
        output['sugars'] = row['sugars']
        output['protein'] = row['protein']
        print(output)
    return
    
df.apply(lambda row: this_question(row), axis = 1) 
df.head()
```

```python
# select all rows with more than and including 400 calories
a = df.where(df['calories'] >= 400)
a.dropna(inplace = True)
a.head()
```

```python
# select all rows whose vitamin c content is higher than the iron content
a = df.where(df['vitamin a'] > df['vitamin c'])
a.dropna(inplace = True)
a.head()
```

```python
# create a new column containing the caffeine per calories of each drink
def this_question(x):
    if x['calories'] != 0:
        cafPerCal = x['caffeine']/x['calories']
        return cafPerCal
    else:
        return
    
df['caffeine per calories'] = df[['caffeine','calories']].apply(lambda row: this_question(row), axis = 1) 
df.head()
```

```python
# what is the average calorie across all items?
import numpy as np

averCals = np.mean(df['calories'])
averCals
```
